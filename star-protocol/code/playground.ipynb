```python
import json

assoc_path = "../outs/associations-generated.jsonl"

# Load the 1001st record (index 1000 is fine, so check 1001)
target_index = 1001

for target_index in range(990, 1020):
    with open(assoc_path) as f:
        for i, line in enumerate(f):
            if i == target_index:
                print("Raw line:", line.strip())
                try:
                    assoc = json.loads(line)
                    print("Parsed:", assoc)
                except Exception as e:
                    print("JSON error:", e)
                break
        else:
            print("No line at index", target_index)
```
